# $r^{3}Y^{3}_\mu(\Omega)$ operator expansion in $SU(3)$ tensors

$$
\left[r^{3}_{\sigma}Y^{3}_\mu(\hat{r}_{\sigma})\right]_{\hspace{7mm}M_{J_o}}^{L_o=J_o=3} 
= \sum_{\eta,l,m_l,\eta',l',m_{l'},m_s,m_{s'}}\Big\langle \eta',l',m_{l'},\frac{1}{2},m_{s'}\Big| r^{3}_{\sigma}Y^{3}_\mu(\hat{r}_{\sigma}) \Big| \eta,l,m_{l},\frac{1}{2},m_{s} \Big\rangle a^{\dagger}_{l',m_{l'},\frac{1}{2},m_{s'}}a_{l,m_l,\frac{1}{2},m_{s}}
\\ = \sum_{\eta,l,m_l,\eta',l',m_{l'},m_s,m_{s'}}
\Big\langle \eta',l',\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle 
\langle l,m_l; 3,\mu | l',m_{l'} \rangle
\langle \frac{1}{2},m_s; 0,0 | \frac{1}{2},m_{s'} \rangle 
a^{\dagger}_{l',m_{l'},\frac{1}{2},m_{s'}}a_{l,m_l,\frac{1}{2},m_{s}}
\\ = \sum_{\eta,l,m_l,\eta',l',m_{l'},m_s,m_{s'}}
\Big\langle \eta',l',\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle 
(-1)^{l-m_l}\frac{\sqrt{2l'+1}}{\sqrt{7}}\langle l',m_{l'}; l,-m_l | 1,\mu \rangle 
(-1)^{\frac{1}{2}-m_s}\sqrt{2}\langle  \frac{1}{2},m_{s'}; \frac{1}{2},-m_s | 0,0  \rangle 
a^{\dagger}_{l',m_{l'},\frac{1}{2},m_{s'}}a_{l,m_l,\frac{1}{2},m_{s}}
\\ = \sum_{\eta,l,m_l,\eta',l',m_{l'},m_s,m_{s'}}
(-1)^{\eta}\sqrt{2(2l'+1)}\frac{\Big\langle \eta',l',\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle}{\sqrt{7}}
\langle l',m_{l'}; l,-m_l | 3,\mu \rangle 
\langle  \frac{1}{2},m_{s'}; \frac{1}{2},-m_s | 0,0  \rangle
a^{\dagger}_{(\eta',0),l',m_{l'},\frac{1}{2},m_{s'}}\tilde{a}_{(0,\eta),l,-m_l,\frac{1}{2},-m_{s}}
\\ = \sum_{\eta,l,\eta',l'}
(-1)^{\eta}\sqrt{\frac{2(2l'+1)}{7}}\Big\langle \eta',l',\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle
\Big\{a^{\dagger}_{(\eta',0),l',\frac{1}{2}}\tilde{a}_{(0,\eta),l,\frac{1}{2}}\Big\}^{L_o=J_o=3}_{\hspace{7mm}M_{J_o}}
\\=\sqrt{\frac{2}{7}}\sum_{(\lambda_o, \mu_o),k_o,\eta,l,\eta',l'}(-1)^{\eta}\sqrt{2l'+1}\Big\langle \eta',l',\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle \langle (\eta',0),0,l'; (0,\eta),0,l||(\lambda_o,\mu_o),k_o,3 \rangle_{\rho_o=1}\\\times\Big\{a^{\dagger}_{(\eta',0)\frac{1}{2}}\times \tilde{a}_{(0,\eta)\frac{1}{2}}\Big\}^{\rho_o=1(\lambda_o,\mu_o)k_oL_o=J_o=3}_{\hspace{26mm}M_{J_o}}
$$

A result to keep in mind in order to check the reduced matrix elements is

$$
\Big\langle \eta',l',\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle = (-1)^{l-l'}\sqrt{\frac{2l+1}{2l'+1}}\Big\langle \eta,l,\frac{1}{2}\Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta',l',\frac{1}{2}\Big\rangle$$

# Program to compute $r^{3}Y^{3}_\mu(\Omega)$ operator expansion in $SU(3)$ tensors for actinides

The expansion is given by

$$\left[r^{3}_{\sigma}Y^{3}_\mu(\hat{r}_{\sigma})\right]_{\hspace{7mm}M_{J_o}}^{L_o=J_o=3} = \sqrt{\frac{2}{7}}\sum_{(\lambda_o,\mu_o)k_o}\sum_{l,l',\eta,\eta'}(-1)^{\eta}\sqrt{2l'+1}\Big\langle \eta',l',\frac{1}{2} \Big|\Big| r^{3}_{\sigma}Y^{3}(\hat{r}_{\sigma}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle\\ \times\langle (\eta',0),0,l'; (0,\eta),0,l||(\lambda_o,\mu_o),k_o,3 \rangle_{\rho_o=1}\Big\{a^{\dagger}_{(\eta',0)\frac{1}{2}}\times \tilde{a}_{(0,\eta)\frac{1}{2}}\Big\}^{\rho_o=1(\lambda_o,\mu_o)k_oL_o=J_o=3}_{\hspace{26mm}M_{J_o}}$$

where $\sigma = \pi, \nu$.

# Proton operator

The expansion is given by 

$$\left[r_{\pi}^3Y^{3}_\mu(\hat{r}_{\pi})\right]_{\hspace{7mm}M_{J_o}}^{L_o=J_o=3} = \sqrt{\frac{2}{7}}\sum_{(\lambda_o,\mu_o)k_o}\sum_{l,l'}\sum_{\eta=5,\eta'=5}^{6}(-1)^{\eta}\sqrt{2l'+1}\Big\langle \eta',l',\frac{1}{2} \Big|\Big| r_{\pi}^3Y^{3}(\hat{r}_{\pi}) \Big|\Big| \eta,l,\frac{1}{2}\Big\rangle \\ \times\langle (\eta',0),0,l'; (0,\eta),0,l||(\lambda_o,\mu_o),k_o,3 \rangle_{\rho_o=1}\Big\{a^{\dagger}_{(\eta',0)\frac{1}{2}}\times \tilde{a}_{(0,\eta)\frac{1}{2}}\Big\}^{\rho_o=1(\lambda_o,\mu_o)k_oL_o=J_o=3}_{\hspace{26mm}M_{J_o}}$$

```python
p_r3Y3 = SU3_rLYL_expansion([5,6], 3)
# (0,1) does not contain L=3
```
