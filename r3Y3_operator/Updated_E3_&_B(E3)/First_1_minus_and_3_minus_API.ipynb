```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import urllib.request
pd.set_option('display.max_columns', None)  


# the service URL
livechart = "https://nds.iaea.org/relnsd/v1/data?"

# There have been cases in which the service returns an HTTP Error 403: Forbidden
def lc_pd_dataframe(url):
    req = urllib.request.Request(url)
    req.add_header('User-Agent', 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:77.0) Gecko/20100101 Firefox/77.0')
    return pd.read_csv(urllib.request.urlopen(req))
```

```python
fields="fields=levels&nuclides=all"  
AllIsotopes = lc_pd_dataframe(livechart + fields)
EvenEvenIsotopes = AllIsotopes[(AllIsotopes["z"]%2==0) &(AllIsotopes["n"]%2==0)]
```

# $E_{1^-_1}$ up to 10 MeV

```python
First1minus = EvenEvenIsotopes.head(0)

for i in range(len(EvenEvenIsotopes)):
    
    # Element to consult
    element = str(EvenEvenIsotopes["symbol"].iloc[i])
    A = str(EvenEvenIsotopes["z"].iloc[i] + EvenEvenIsotopes["n"].iloc[i])
    fields="fields=levels&nuclides="+A+element  

    IsotopeLevels = lc_pd_dataframe(livechart + fields)
    
    # Filters over 1-, (1-), 1(-) and (1)- values
    IsotopeLevels_1_minus = IsotopeLevels[(IsotopeLevels["jp"] == "1-") | (IsotopeLevels["jp"] == "(1-)") | (IsotopeLevels["jp"] == "1(-)") | (IsotopeLevels["jp"] == "(1)-")]    
    IsotopeLevels_1_minus_lowest = IsotopeLevels_1_minus.sort_values(by='energy').head(1)
    First1minus = First1minus.append(IsotopeLevels_1_minus_lowest)
```

```python
First1minusBelow = First1minus[First1minus["energy"]<=10000]
# Probably a wrong value of 6.71 eV
First1minusBelow = First1minusBelow[(First1minusBelow["n"] != 114) & ((First1minusBelow["z"] != 76) ) ]

First1minusBelow.sort_values(by=["energy"]).head(12)
# Lowest values: 224Ra, 226Th, 222Ra, 222Th, 224Th (Quartet), 226Ra, 228Th, 230U, 220Ra (Quartet), 228Ra, 230Th, 240Pu
```

```python

#%matplotlib notebook

# Data
N = First1minusBelow["n"].to_numpy()
Z = First1minusBelow["z"].to_numpy()
E = First1minusBelow["energy"].to_numpy()
magic = [2, 8, 20, 28, 50, 82, 126]

# Canvas
fig = plt.figure(figsize = (15,15))
ax = plt.axes(projection ="3d")
 
#Plots
ax.scatter(N, Z, E, "o", s= 10, c = "C0", alpha=1)
ax.scatter(N, Z, 0, "o", s= 10, c = "C0", alpha=0.5)
ax.scatter(N, max(Z), E, "o", s= 10, c = "C1", alpha=1)
ax.scatter(min(N), Z, E, "o", s= 10, c = "C2", alpha=1)
for i in range(len(N)):
    ax.plot([N[i], N[i]], [Z[i], Z[i]], [E[i], 0], color='C0', alpha=0.1)  
#for i in range(len(magic)):
    #ax.plot([magic[i], magic[i]], [105, 105], [0, max(E)], color='C1', alpha=0.4)  
    #if i < len(magic)-1:
    #    ax.plot([0,0], [magic[i], magic[i]], [0, max(E)], color='C2', alpha=0.4)  

    
# Format the axis
ax.invert_xaxis()
ax.set_xlabel('N', fontsize=14, labelpad=10)
ax.set_ylabel('Z', fontsize=14, labelpad=10)
ax.set_zlabel(r'$E(1^{-}_1)$ [keV]', fontsize=14, labelpad=10, rotation=-180)
ax.view_init(35, -135) 
ax.set_xticks(np.arange(min(N), max(N)+10, 10))
ax.set_yticks(np.arange(min(Z), max(Z)+10, 10))
ax.set_zticks(np.arange(300, 9600, 1000))

# Plot the planes

maxE = 10000

X, Y = np.meshgrid(np.linspace(min(N), max(N), 10), np.linspace(min(Z), max(Z), 10))
Z_xy = np.zeros_like(X)
X_xz, Z_xz = np.meshgrid(np.linspace(min(N), max(N), 10), np.linspace(0, maxE, 10))
Y_xz = np.full((len(X_xz[0]), len(X_xz[0])), max(Z))
Y_yz, Z_yz = np.meshgrid(np.linspace(min(Z), max(Z), 10), np.linspace(0, maxE, 10))
X_yz = np.full((len(X_xz[0]), len(X_xz[0])), min(N))
ax.plot_surface(X, Y, Z_xy, color='blue', alpha=0.05, rstride=100, cstride=100)
ax.plot_surface(X_xz, Y_xz, Z_xz, color='red', alpha=0.05, rstride=100, cstride=100)
ax.plot_surface(X_yz, Y_yz, Z_yz, color='green', alpha=0.05, rstride=100, cstride=100)
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

# Maximum values
ax.set_xlim(max(N), min(N))
ax.set_ylim(min(Z), max(Z))
ax.set_zlim(min(E), maxE)

plt.savefig('E1minus.png', format='png', dpi=800)
plt.show()
```
